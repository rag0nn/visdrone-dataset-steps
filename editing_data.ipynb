```python
import os
import numpy as np
```

```python
main_file_names =["VisDrone2019-DET-test-dev","VisDrone2019-DET-train","VisDrone2019-DET-val"]
sub_title = ["labels"]
```

```python
for file in main_file_names:
    for title in sub_title:
        txt_files = os.listdir(f"./{file}/{title}")
        
        for txt_file in txt_files:
            txt_file_path = f"./{file}/{title}/{txt_file}"
            print(f"processing: {txt_file_path}")
            # read pre tags
            f = open(txt_file_path,"r")
            lines = f.readlines()
            f.close()
            
            # write new tags
            f = open(txt_file_path,"w")
            for line in lines:
                new_line = convert_tag(line)
                f.writelines(new_line)
            f.close()
```

```python
target_tags = {
  0: 11,
  1: 11,
  2: 11,
  3: 0,
  4: 2,
  5: 4,
  6: 1,
  7: 1,
  8: 2,
  9: 1
}

def convert_tag(line):
    tag = line.split(" ")[0]
    new_tag = target_tags[int(tag)]
    return f"{new_tag}{line[1:]}"
```

### Değerlere bakalım

```python
counts=  {
    1:0,
    0:0,
    2:0,
    4:0,
    11:0
}

for file in main_file_names:
    for title in sub_title:
        txt_files = os.listdir(f"./{file}/{title}")
        
        for txt_file in txt_files:
            txt_file_path = f"./{file}/{title}/{txt_file}"
            #print(f"processing: {txt_file_path}")
            # read pre tags
            f = open(txt_file_path,"r")
            lines = f.readlines()
            f.close()
            
            for line in lines:
                tag = int(line.split(" ")[0])
                counts[tag] +=1


import pandas as pd
pd.DataFrame(counts.values(),index=counts.keys())
```
